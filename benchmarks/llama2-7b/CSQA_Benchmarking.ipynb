```python
from datasets import load_dataset
import os

destination_dir = '/content/opencompass/data/commonsenseqa'

os.makedirs(destination_dir, exist_ok=True)

ds = load_dataset("tau/commonsense_qa")

for split in ds.keys():
    save_path = os.path.join(destination_dir, f"{split}.jsonl")
    ds[split].to_json(save_path)
    print(f"Saved {split} split to {save_path}.")

print(f"Dataset saved successfully to {destination_dir}.")
```
